```python
class Board:
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        self.board = [['O' for _ in range(cols)] for _ in range(rows)]

    def play(self, col, color):
        if col < 0 or col >= self.cols:
            print("Invalid column. Choose a column between 0 and", self.cols - 1)
            return False

        for row in range(self.rows - 1, -1, -1):
            if self.board[row][col] == 'O':
                self.board[row][col] = color
                return True

        print("Column is full. Choose another column.")
        return False

    def print_board(self):
        for row in self.board:
            print(' '.join(row))
        print('-' * (self.cols * 2 - 1))
        print(' '.join(str(i) for i in range(self.cols)))

def check_winner(board, color):
    # Check for horizontal win
    for row in range(board.rows):
        for col in range(board.cols - 3):
            if board.board[row][col] == board.board[row][col + 1] == board.board[row][col + 2] == board.board[row][col + 3] == color:
                return True

    # Check for vertical win
    for col in range(board.cols):
        for row in range(board.rows - 3):
            if board.board[row][col] == board.board[row + 1][col] == board.board[row + 2][col] == board.board[row + 3][col] == color:
                return True

    # Check for diagonal win (bottom-left to top-right)
    for row in range(3, board.rows):
        for col in range(board.cols - 3):
            if board.board[row][col] == board.board[row - 1][col + 1] == board.board[row - 2][col + 2] == board.board[row - 3][col + 3] == color:
                return True

    # Check for diagonal win (bottom-right to top-left)
    for row in range(3, board.rows):
        for col in range(3, board.cols):
            if board.board[row][col] == board.board[row - 1][col - 1] == board.board[row - 2][col - 2] == board.board[row - 3][col - 3] == color:
                return True

    return False

def main():
    rows = 6
    cols = 7
    board = Board(rows, cols)
    current_player = 'R'

    while True:
        board.print_board()
        col = int(input(f"Player {current_player}, choose a column (0-{cols - 1}): "))

        if board.play(col, current_player):
            if check_winner(board, current_player):
                board.print_board()
                print(f"Player {current_player} wins! You receive 100,000 euros!")
                break

            current_player = 'J' if current_player == 'R' else 'R'

if __name__ == "__main__":
    main()
```
